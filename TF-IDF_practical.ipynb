```python
import pandas as pd
messages = pd.read_csv('SPMclassifer\SMSSpamCollection.txt',sep='\t', names = ['label','message'])
```

```python
messages
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]'," ",messages['message'][i])
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review ]
    review = ' '.join(review)
    corpus.append(review)
```

## create TF-IDF and NGrams

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=100,lowercase=True,stop_words=stopwords.words('english'))
X = tfidf.fit_transform(corpus).toarray()
```

```python
tfidf = TfidfVectorizer(max_features=100,lowercase=True,stop_words=stopwords.words('english'),ngram_range=(2,3))
tfidf.fit_transform(corpus).toarray()
```
