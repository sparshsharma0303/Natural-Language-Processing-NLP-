# Stemming
* text-processing technique that reduces words to their base or root form by removing prefixes and suffixes

* standardizes words which helps to improve the efficiency and effectiveness of various natural language processing (NLP) tasks.

* example:
    chocolates ---> chocolate
    retrieva ---> retrive
    

## Types of Stemming


1. Regexp Stemmer
* flexible stemming algorithm that allows users to define custom rules using regular expressions (regex). 
* This stemmer can be helpful for very specific tasks where predefined rules are necessary for stemming.
* Example:
'running' → 'runn'
Custom rule: r'ing$' removes the suffix ing.




```python
from nltk.stem import RegexpStemmer

reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min = 4)
reg_stemmer.stem('eating')
```

3. Snowball Stemmer
* enhanced version of the Porter Stemmer 
* is faster and more aggressive than its predecessor. One of the key advantages of this is that it supports multiple languages, making it a multilingual stemmer.


```python
words = ["eating","eats", "eaten","writing","writes","programming","programs","history","finally", "finalized" ]
```

```python
from nltk.stem import SnowballStemmer

snowballstemmer = SnowballStemmer('english')

for word in words:
    print(f"{word}------>{snowballstemmer.stem(word)}")
```

major disadvantage of stemming is that some words are changed to completely meaning less words


2. Porter Stemmer
* works by applying a series of rules to remove common suffixes from English words.  stemmed output is not guaranteed to be a meaningful word and its applications are limited to the English language.

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
for word in words:
    print(f"{word} -----> {stemming.stem(word)}")
    
```

```python
stemming.stem('Fairly'),stemming.stem('sportingly')
```

```python
snowballstemmer.stem('Fairly'),snowballstemmer.stem('sportingly')
```

### Lemmatization

output will get after lemmatization is called "lemma", which is a root word, rather than root stem 


```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
'''
POS-
Noun-n
verb -v
adjective -a
adverb -r
'''
lemmatizer.lemmatize('going',pos='v')
```

```python
for word in words:
    print(word+"------->"+lemmatizer.lemmatize(word))
```

```python
lemmatizer.lemmatize("goes" ,pos ='v')
```

```python
lemmatizer.lemmatize("sportingly" ,pos ='n'),lemmatizer.lemmatize("fairly" ,pos ='v')
```
